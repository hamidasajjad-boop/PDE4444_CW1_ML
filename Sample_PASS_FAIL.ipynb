```python
import matplotlib.pyplot as plt
from PIL import Image
import os, random

def show_samples(folder, label, n=8):
    paths = random.sample([os.path.join(folder, f) 
                           for f in os.listdir(folder)], n)
    fig, axes = plt.subplots(1, n, figsize=(20, 3))
    fig.suptitle(label, fontsize=14, fontweight='bold')
    for ax, p in zip(axes, paths):
        ax.imshow(Image.open(p))
        ax.axis('off')
    plt.tight_layout()
    plt.show()

base = r"C:\Users\erum8\barcode_dataset_with_fail\train"
show_samples(os.path.join(base, "PASS"), "✅ PASS samples")
show_samples(os.path.join(base, "FAIL"), "❌ FAIL samples")
```
